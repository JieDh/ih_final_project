```python
import pandas as pd
import re
```

```python
s1 = pd.read_csv('clean_data/cleansers.csv')
s2 = pd.read_csv('clean_data/creams.csv')
s3 = pd.read_csv('clean_data/essence.csv')
s4 = pd.read_csv('clean_data/eyecreams.csv')
s5 = pd.read_csv('clean_data/masks.csv')
s6 = pd.read_csv('clean_data/scrubs.csv')
s7 = pd.read_csv('clean_data/sunscreens.csv')
s8 = pd.read_csv('clean_data/toner.csv')
```

```python
df = pd.concat([s1, s2, s3, s4, s5, s6, s7, s8], ignore_index=True)
```

```python
df
```

# 1. Separar marca y producto

```python
df['name'] = df['name'].apply(lambda x: x.split(' - '))
```

```python
df['brand'] = df['name'].apply(lambda x: x[0])
df['product name'] = df['name'].apply(lambda x: x[1:])
```

```python
df
#el nombre del producto contiene más info de la necesaria, corregir
```

# 2. Separar producto y tamaño

```python
product = []
for i in df['product name']:
    try:
        product.append(i[0])
    except:
        product.append(i)
print(len(product))
```

```python
df['product'] = product
```

```python
df
```

# 3. Separar detalles

```python
df['details'] = df['details'].astype(str)
```

```python
df['details'] = df['details'].apply(lambda x: x.split('\n\n\n'))
```

```python
df
```

```python
product_type = []
for x in df['details']:
    product_type.append(re.findall(r"(Product Type: [a-zA-Z].+?)'", str(x)))
```

```python
pr_type = []
for x in product_type:
    if len(x) >= 1:
        for i in x:
            pr_type.append(i)
    else:
        pr_type.append(x)
df['pr_type'] = pr_type
df['pr_type'] = df['pr_type'].apply(lambda x: str(x).split(':'))
```

```python
product = []
for i in df['pr_type']:
    try:
        product.append(i[1])
    except:
        product.append(i)
print(len(product))
```

```python
df['product type'] = product
```

```python
skin_type = []
for x in df['details']:
    skin_type.append(re.findall(r"(Skin Type: [a-zA-Z].+?)'", str(x)))
sk_type = []
for x in skin_type:
    if len(x) >= 1:
        for i in x:
            sk_type.append(i)
    else:
        sk_type.append(x)
df['sk_type'] = sk_type
df['sk_type'] = df['sk_type'].apply(lambda x: str(x).split(':'))
skin_t = []
for i in df['sk_type']:
    try:
        skin_t.append(i[1])
    except:
        skin_t.append(i)
print(len(skin_t))
```

```python
df['skin type'] = skin_t
```

```python
skin_concern = []
for x in df['details']:
    skin_concern.append(re.findall(r"(Skin Concern: [a-zA-Z].+?)'", str(x)))
sk_concern = []
for x in skin_concern:
    if len(x) >= 1:
        for i in x:
            sk_concern.append(i)
    else:
        sk_concern.append(x)
df['sk_concern'] = sk_concern
df['sk_concern'] = df['sk_concern'].apply(lambda x: str(x).split(':'))
skin_c = []
for i in df['sk_concern']:
    try:
        skin_c.append(i[1])
    except:
        skin_c.append(i)
print(len(skin_c))
df['skin concern'] = skin_c
```

# 4. Droppear columnas no necesarias

```python
df.drop(columns=['name', 'Unnamed: 0', 'Unnamed: 0.1', 'details', 'product name', 'pr_type', 'sk_type', 'sk_concern'], axis=1, inplace=True)
```
